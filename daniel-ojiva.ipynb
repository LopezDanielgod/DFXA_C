# 3.Ojiva

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Datos del ejemplo ---
data = [15, 22, 18, 25, 20, 23, 19, 21, 26, 17, 24, 20, 22, 16, 28, 21, 19, 23, 20, 25, 18, 26, 17, 24, 22, 20, 21, 19, 27, 23]
data.sort() # Ordenar los datos es crucial para las frecuencias acumuladas

# --- 2. Definir los intervalos de clase y calcular frecuencias ---
# Definimos los límites de los intervalos.
# Aquí empezamos desde 14.5 para el primer punto y terminamos un poco después del máximo.
# Los intervalos son de 3 minutos de ancho.
bins = np.arange(14.5, max(data) + 3.5, 3)

# Calcular las frecuencias absolutas (histograma)
hist, bin_edges = np.histogram(data, bins=bins)

# Calcular las frecuencias acumuladas
# Añadimos un 0 al principio para que la ojiva comience desde el eje X
cumulative_frequencies = np.concatenate(([0], np.cumsum(hist)))

# Los puntos X para la ojiva son los límites superiores de cada bin
# (excluyendo el primero que usamos para el 0 de la frecuencia acumulada)
# y luego se añade el último límite superior.
x_values_ojiva = bin_edges

# --- 3. Crear la gráfica de ojiva ---
plt.figure(figsize=(10, 6)) # Tamaño de la figura para mejor visualización
plt.plot(x_values_ojiva, cumulative_frequencies, marker='o', linestyle='-', color='skyblue', label='Frecuencia Acumulada')

# Etiquetas y título
plt.title('Gráfica de Ojiva (Tiempos de Examen)', fontsize=16)
plt.xlabel('Tiempo (minutos)', fontsize=12)
plt.ylabel('Número de Estudiantes (Frecuencia Acumulada)', fontsize=12)

# Añadir una cuadrícula para facilitar la lectura
plt.grid(True, linestyle='--', alpha=0.7)

# Añadir marcas en los ejes para cada intervalo y frecuencia
plt.xticks(bins) # Marcas en los límites de los intervalos
plt.yticks(np.arange(0, len(data) + 1, 5)) # Marcas cada 5 estudiantes

# Añadir texto a los puntos de datos para mayor claridad
for i, txt in enumerate(cumulative_frequencies):
    plt.annotate(txt, (x_values_ojiva[i], cumulative_frequencies[i]), textcoords="offset points", xytext=(0,5), ha='center')

# Mostrar la leyenda
plt.legend()

# Mostrar la gráfica
plt.show()

# --- Explicación Adicional (Opcional) ---
print("\n--- Detalles de la tabla de frecuencias ---")
print(f"Límites de los bins: {bins}")
print(f"Frecuencias absolutas por bin: {hist}")
print(f"Frecuencias acumuladas: {cumulative_frequencies}")
```
