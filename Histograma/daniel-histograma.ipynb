# 1. Histograma

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Datos del ejemplo (edades) ---
ages = [23, 35, 42, 28, 51, 30, 25, 48, 33, 39, 22, 55, 40, 27, 31, 45, 36, 29, 50, 24,
        38, 43, 32, 26, 47, 34, 41, 37, 49, 21, 52, 44, 30, 28, 39, 46, 33, 25, 53, 40,
        22, 54, 31, 29, 42, 36, 27, 48, 35, 23]

# --- 2. Definir los intervalos de clase (bins) ---
# np.arange(inicio, fin + paso, paso)
# Empieza en 20, va hasta 60 (para incluir el 55), con un paso de 5.
bins = np.arange(20, 60, 5)

# --- 3. Crear la gráfica del histograma ---
plt.figure(figsize=(10, 6)) # Tamaño de la figura

# plt.hist() es la función para crear histogramas.
# 'edgescolor' hace que las barras tengan un borde para distinguirlas mejor.
plt.hist(ages, bins=bins, edgecolor='black', color='lightcoral', alpha=0.7)

# Etiquetas y título
plt.title('Histograma de Edades de Participantes', fontsize=16)
plt.xlabel('Edad (años)', fontsize=12)
plt.ylabel('Número de Participantes', fontsize=12)

# Añadir una cuadrícula para facilitar la lectura
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Marcas en el eje X para cada bin
plt.xticks(bins)

# Opcional: Mostrar los conteos exactos en la parte superior de cada barra
# Esto requiere un poco más de código para extraer las alturas de las barras
n, bins, patches = plt.hist(ages, bins=bins, edgecolor='black', color='lightcoral', alpha=0.7)
for i in range(len(patches)):
    plt.text(patches[i].get_x() + patches[i].get_width() / 2, patches[i].get_height(),
             int(n[i]), ha='center', va='bottom')

# Mostrar la gráfica
plt.show()

# --- Detalles de los Bins y Frecuencias (opcional) ---
print("\n--- Detalles de los bins y frecuencias ---")
counts, _ = np.histogram(ages, bins=bins)
print(f"Límites de los bins: {bins}")
print(f"Frecuencias por bin: {counts}")
```
