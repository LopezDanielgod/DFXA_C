```python
import matplotlib.pyplot as plt

# Configuracin del tamao de la figura
plt.figure(figsize=(12, 6))  # (Ancho, Alto) en pulgadas

# Datos
marcas_clase = [1, 2, 3, 4, 5, 6]  # Eje x (marcas de clase como ndices)
frecuencia = [0.05, 0.15, 0.20, 0.20, 0.10, 0.30]  # Eje y (frecuencias relativas)

# Ajuste para el polgono de frecuencias (aadir extremos en cero)
datos_x = [0] + marcas_clase + [7]
datos_y = [0] + frecuencia + [0]

# Etiquetas de texto para marcas de clase
marcas_textos = ["Coca", "Toronja", "Sprite", "Manzanita", "Punch", "Jarrito de pia"]

# Crear el grfico del polgono de frecuencias
plt.plot(datos_x, datos_y,
         linewidth=5,
         color="r",
         linestyle=":",            # Lnea punteada
         marker="v",               # Marcador triangular hacia abajo
         markersize=15,
         markeredgecolor="b",     # Borde del marcador azul
         markerfacecolor="y")     # Relleno del marcador amarillo

# Personalizacin de ejes
plt.xticks(marcas_clase, marcas_textos, fontsize=15, rotation=45)
plt.xlabel("Marcas de clase", fontsize=25)
plt.ylabel("Frecuencia relativa", fontsize=25)
plt.title("Polgono de Frecuencias", fontsize=40)
plt.grid(True)

# Mostrar el grfico
plt.show()

print("Setup terminado")

```
