```python
import matplotlib.pyplot as plt
import seaborn as sns

labels = ["Coca-Cola", "Pepsi", "Sprite", "Fanta"]
values = [80, 60, 40, 50]
colors = ["#e60023", "#0033A0", "#77C043", "#FFA500"]

plt.figure(figsize=(8, 5))
sns.barplot(x=values, y=labels, palette=colors)

plt.xlabel("Consumo (%)")
plt.ylabel("Refresco")
plt.title("Consumo de Refrescos Populares")
plt.xticks(range(0, 101, 20))  

for i, v in enumerate(values):
    plt.text(v + 2, i, f"{v}%", color="black", va="center", fontweight="bold")

plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```
