## HISTOGRAMA

```python
import matplotlib.pyplot as plt
print("Setup terminado")
```

```python
plt.figure(figsize = (12, 6)) # (Ancho,Alto) de graficas

marcas_clase = [0, 1, 2, 3, 4, 5] # Datos del eje x
frecuencia = [5, 15, 20, 20, 10, 30] # Datos del eje y
marcas_textos = ["Coca","Toronja", "Sprite", "Manzanita", "Punch", "Kosako"]

# Histograma con: # - barras ajustadas al 100% (widh = 1)
# - Contorno negro (edgcolor = "k")
plt.bar(marcas_clase , frecuencia, width = 1, edgecolor = "k",
        color = ["#ED4C29", "#F9A704", "#FAE408", "#81FA08", "#03F78F", "#03CAF7"])

plt.xticks(marcas_clase, marcas_textos, fontsize = 15, rotation = 45)
plt.xlabel("Marcas de clase", fontsize = 25) # Etiqueta en eje x
plt.ylabel("Frecuencia absoluta", fontsize = 25) # Etiqueta en eje y
plt.title("Histograma", fontsize = 40) # Titulo
plt.grid() # Se activa la cuadricula
plt.show() # Se muestra en el grafico en pantallas
```
