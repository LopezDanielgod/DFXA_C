## DIAGRAMA DE PASTEL

```python
import matplotlib.pyplot as plt
print("Setup terminado")
```

```python
datos = [5, 12, 23, 25, 20, 15]
separaciones = [0, 0, 0.2, 0, 0, 0]

plt.pie(datos,
       explode = separaciones,
       labels = ["Coca","Toronja", "Sprite", "Manzanita", "Kosako", "Jarrito de piña"],
       colors = ["#ED4C29", "#F9A704", "#FAE408", "#81FA08", "#03F78F", "#03CAF7"],
       autopct = "%0.01f%%",
       pctdistance = 0.8,
       counterclock = False,
       startangle = 90)
plt.show()
```
