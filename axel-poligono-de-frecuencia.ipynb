```python
import matplotlib.pyplot as plt

# Datos
puntos_medios = [4.5, 14.5, 24.5, 34.5, 44.5]  # Marcas de clase
frecuencias = [2, 5, 8, 6, 3]  # Frecuencias absolutas

# Cierre visual del polgono (opcional)
datos_x = [0] + puntos_medios + [50]
datos_y = [0] + frecuencias + [0]

# Graficar el polgono
plt.figure(figsize=(10, 6))
plt.plot(datos_x, datos_y,
         marker='o',
         linestyle='-',
         color='green',
         linewidth=3,
         markersize=10,
         markerfacecolor='orange')

# Personalizacin
plt.title("Polgono de Frecuencias", fontsize=20)
plt.xlabel("Punto Medio (Edad)", fontsize=15)
plt.ylabel("Frecuencia", fontsize=15)
plt.grid(True)

# Etiquetas personalizadas (opcionales)
plt.xticks(puntos_medios)

plt.show()

```
